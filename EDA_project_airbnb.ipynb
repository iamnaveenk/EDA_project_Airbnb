**4. Is there any noticeable difference of traffic among different areas and what could be the reason for it?**

```python
traffic_areas2= data.groupby(['room_type','neighbourhood_group'])['minimum_nights'].count().reset_index()
traffic_areas2.sort_values('minimum_nights',axis=0,ascending=False).head()
```

```python
traffic_areas3= data.groupby(['room_type','neighbourhood_group'])['minimum_nights'].mean().reset_index()
traffic_areas3.sort_values('minimum_nights',axis=0,ascending=False).head()
```

```python
traffic_areas3= data.groupby(['room_type','neighbourhood_group'])['price'].mean().reset_index()
traffic_areas3.sort_values('price',axis=0,ascending=False).head()
```

```python
#creating the bar plot
x=traffic_areas2['room_type']
y=traffic_areas2['minimum_nights']

plt.bar(x,y,color='blue',width=0.5)
plt.xlabel("Room type")
plt.ylabel("minimum number of night stayed")
plt.title("traffic Areas")
plt.show()
```

```python
x=traffic_areas2['neighbourhood_group']
y=traffic_areas2['minimum_nights']

plt.bar(x,y,color='blue',width=0.5)
plt.xlabel("neighbourhood_group")
plt.ylabel("minimum number of night stayed")
plt.title("traffic Areas")
plt.show()
```

**OBSERVATION:** As we can see the most traffic area which is in room_type is entire home/apt,follwed by private room and we can also say that shared room is less preffered and in

```python
reviews_prediction1=data.groupby(['room_type','neighbourhood_group','neighbourhood','price','minimum_nights'])['number_of_reviews'].count().reset_index()

reviews_prediction1.sort_values('number_of_reviews', axis=0, ascending=False).head()
```

```python
x=reviews_prediction1['minimum_nights']
y=reviews_prediction1['price']
fig = plt.figure(figsize = (10, 5))
plt.scatter(x, y)

plt.xlabel("minimum_nights")
plt.ylabel("price")
plt.title("price Vs minimum_nights")
plt.show()
```
