## <b> Who is AIRBNB? <b>

AIRBNB is an American company, founded in august 2008 and based on San Francisco , california.

## <b>  How it operates ? <b>

Airbnb operates on online market place for lodging, homestays for vacation rental or provide homestay for holidays spednding people and tourism activities.

Airbnd does not have own properties , they have platform (website and mob App) people come to thier platform search property as per requirement and book the property , their profit comes from commision of each booking . 

## <b> Since 2008, guests and hosts have used Airbnb to expand on traveling possibilities and present a more unique, personalized way of experiencing the world. Today, Airbnb became one of a kind service that is used and recognized by the whole world. Data analysis on millions of listings provided through Airbnb is a crucial factor for the company. These millions of listings generate a lot of data - data that can be analyzed and used for security, business decisions, understanding of customers' and providers' (hosts) behavior and performance on the platform, guiding marketing initiatives, implementation of innovative additional services and much more. </b>

**1.What can we learn about different hosts and areas?**

```python
host_areas= df.groupby(['host_name','neighbourhood_group'])['calculated_host_listings_count'].count().reset_index()
```

```python
host_areas.head(5)
```

```python
host_areas.sort_values("calculated_host_listings_count",ascending=False).head().rename(columns={'host_name':'most active host','neighbourhood_group': 'location', 'calculated_host_listings_count':'hosting'})
```

```python
x=host_areas['neighbourhood_group']
y=host_areas['calculated_host_listings_count']

fig=plt.figure(figsize=(10,5))

plt.bar(x,y,color='blue',width=0.5)
plt.xlabel("neighbourhood_group")
plt.ylabel("calculated_host_listings_count")
plt.title("neighbourhood_group vs calculated_host_listings_count")
plt.show()
```

**Observation:**
As we can see host Sounder(NYC) is most potential  host name followed by Blueground, Michael , David and they all are choose Manhattan .
so we can say that Manhattan is mostly preffered location by host.

```python
host_roomtype= df.groupby(['host_name','room_type','neighbourhood_group'])['calculated_host_listings_count'].count().reset_index()
```

```python
host_roomtype.sort_values('calculated_host_listings_count',ascending=False).head(5)
```

```python
fig=plt.figure(figsize=(10,7))
df.groupby(['room_type'])['calculated_host_listings_count'].count().plot(kind='pie', y='calculated_host_listings_count',title='most preffered room type',autopct='%1.1f%%', colors=['green','blue','pink'],radius=1.3,textprops={'fontsize': 20})
```

**observation:**

```python
x=host_roomtype['room_type']
y=host_roomtype['calculated_host_listings_count']

fig=plt.figure(figsize=(15,7))

plt.bar(x,y,color='blue',width=0.5)
plt.xlabel("room_type")
plt.ylabel("calculated_host_listings_count")
plt.title("room_type vs calculated_host_listings_count")
plt.show()
```

```python
reviews_prediction1=df.groupby(['room_type','neighbourhood_group','neighbourhood','price','minimum_nights'])['number_of_reviews'].count().reset_index()

reviews_prediction1.sort_values('number_of_reviews', axis=0, ascending=False).head()
```

**Observation:**
As we can see people are give maximum reviews where price is less

**Obervation:**
As we can see from above where price is less maximum people are spent maximum nights.

```python
reviews_prediction2=df.groupby(['room_type','neighbourhood_group','neighbourhood','number_of_reviews','minimum_nights'])['price'].mean().reset_index()

reviews_prediction2.sort_values('price', axis=0, ascending=False).head()
```
